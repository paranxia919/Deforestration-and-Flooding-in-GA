```python
merged_details = pd.merge(details, location[['EVENT_ID', 'LOCATION', 'LATITUDE', 'LONGITUDE']], on='EVENT_ID', how='left')
merged_details = merged_details[['EVENT_ID', 'STATE', 'EVENT_TYPE', 'BEGIN_YEARMONTH', 'END_YEARMONTH', 'BEGIN_TIME', 'END_TIME','BEGIN_LAT', "BEGIN_LON", "END_LAT", "END_LON", "LOCATION", "LATITUDE", "LONGITUDE"]]
merged_details
```

```python
merged_details = merged_details[merged_details['STATE'] == 'GEORGIA']
merged_details = merged_details[
    (merged_details['EVENT_TYPE'] == 'Flash Flood') |
    (merged_details['EVENT_TYPE'] == 'Coastal Flood') |
    (merged_details['EVENT_TYPE'] == 'Flood')
]
merged_details = merged_details.dropna()
merged_details
```

```python
jan24 = merged_details[merged_details['BEGIN_YEARMONTH'] == 202401]
jan24
```

```python
Feb24 = merged_details[merged_details['BEGIN_YEARMONTH'] == 202402]
Feb24
```

```python
latitudes = Feb24['LATITUDE']
longitudes = Feb24['LONGITUDE']
m2 = folium.Map(location=[latitudes.iloc[0], longitudes.iloc[0]], zoom_start=10)

# Add markers for each location in the dataset
for lat, lon in zip(latitudes, longitudes):
    folium.Marker([lat, lon]).add_to(m2)

# Save the map as an HTML file to view interactively
m2.save('map.html')
m2
```
